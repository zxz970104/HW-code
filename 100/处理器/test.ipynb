```python
import copy
res = []

# 将可选的处理器组合都找出
def dfs(link, index, num, path):
    if len(path) == num:
        res.append(copy.deepcopy(path))
        return
    for i in range(index, len(link)):
        path.append(link[i]);
        dfs(link, i+1, num, path)
        path.pop()

def getResult(arr, num):
    link1 = []
    link2 = []
    for i in range(len(arr)):
        if arr[i] < 4:
            link1.append(arr[i])
        else:
            link2.append(arr[i])
    len1 = len(link1)
    len2 = len(link2)
    if num == 1: # 1 3 2 4
        if len1 == 1 or len2 == 1:
            if len1 == 1:
                dfs(link1, 0, num, [])
            if len2 == 1:
                dfs(link2, 0, num, [])
        elif len1 == 3 or len2 == 3:
            if len1 == 3:
                dfs(link1, 0, num, [])
            if len2 == 3:
                dfs(link2, 0, num, [])
        elif len1 == 2 or len2 == 2:
            if len1 == 2:
                dfs(link1, 0, num, [])
            if len2 == 2:
                dfs(link2, 0, num, [])
        elif len1 == 4 or len2 == 4:
            if len1 == 4:
                dfs(link1, 0, num, [])
            if len2 == 4:
                dfs(link2, 0, num, [])

    elif num == 2: # 2 4 3
        if len1 == 2 or len2 == 2:
            if len1 == 2:
                dfs(link1, 0, num, [])
            if len2 == 2:
                dfs(link2, 0, num, [])
        elif len1 == 4 or len2 == 4:
            if len1 == 4:
                dfs(link1, 0, num, [])
            if len2 == 4:
                dfs(link2, 0, num, [])
        elif len1 == 3 or len2 == 3:
            if len1 == 3:
                dfs(link1, 0, num, [])
            if len2 == 3:
                dfs(link2, 0, num, [])
    
    elif num == 4: # 2 4 3
        if len1 == 4 or len2 == 4:
            if len1 == 4:
                dfs(link1, 0, num, [])
            if len2 == 4:
                dfs(link2, 0, num, [])

    elif num == 8:
        if len1 == 4 and len2 == 4:
            tmp = []
            tmp.extend(link1) # 合并列表
            tmp.extend(link2)
            res.append(tmp)

line = input()
line = line[1:len(line)-1]
print(line)
arr = [int(x) for x in line.split(',')]
num = int(input())

getResult(arr, num)
print(res)

         
```
